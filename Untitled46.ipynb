```python
import json
import numpy as np

with open('data.json') as f:
    data = json.load(f)

ages = [d['patient_age'] for d in data]
num_medicines = [d['num_medicines'] for d in data]


corr_coef = np.corrcoef(ages, num_medicines)[0, 1]


corr_coef = round(corr_coef, 2)

print(corr_coef)
```
